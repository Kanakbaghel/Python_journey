## <font size = 6 color = steelblue ><b>Merging & Concatenation

````python
pers = pd.DataFrame({'EnrolNo': ['A101', 'A102', 'A109', 'A107', 'A110', 'A112'],
                    'Names': ['Aman', 'Shreay', 'Paul', 'Sam', 'John', 'Kia'],
                    'Interest' : ['Music', 'Arts', 'Music', 'Arts', 'Drama', 'Arts'],
                    'Age' : [12,13,12, 13,13, 12]})

marks = pd.DataFrame({'EnrolNo': ['A001', 'A102', 'A105', 'A109', 'A110'],
                     'English':[33, 44, 35, 37, 49],
                     'Math': [44,41,32, 30, 43] })
````

````python
pd.merge(pers, marks) # inner merge
````

### <font size = 5 color = salmon ><b> inner merge : includes only matching rows

````python
inner_merge = pd.merge(pers, marks, how = 'inner', on = 'EnrolNo', indicator= True)
inner_merge
````

### <font size = 5 color = salmon ><b> outer merge: includes all rows

````python
outer_merge = pd.merge(pers, marks, how = 'outer', on = 'EnrolNo', indicator= True)
outer_merge
````

### <font size = 5 color = salmon ><b> left merge: rows from only 1st df ( left df )

````python
left_merge = pd.merge(pers, marks, how = 'left', on = 'EnrolNo', indicator= True)
left_merge
````

### <font size = 5 color = salmon ><b> right merge: rows from only 2nd df ( right df )

````python
right_merge = pd.merge(pers, marks, how = 'right', on = 'EnrolNo', indicator= True)
right_merge
````

#### more than 1 common column

````python
pers = pd.DataFrame({'EnrolNo': ['A101', 'A102', 'A109', 'A107', 'A110', 'A112'],
                    'Names': ['Aman', 'Shreay', 'Paul', 'Sam', 'John', 'Kia'],
                    'Interest' : ['Music', 'Arts', 'Music', 'Arts', 'Drama', 'Arts'],
                    'Age' : [12,13,12, 13,13, 12]})

marks = pd.DataFrame({'EnrolNo': ['A001', 'A102', 'A105', 'A109', 'A110'],
                      'Names' : ['Amy', 'Shreya', 'Mary', 'Paul', 'John'],
                     'English':[33, 44, 35, 37, 49],
                     'Math': [44,41,32, 30, 43] })
````

````python
pd.merge(pers, marks)
````

````python
merged = pd.merge(pers, marks, how = 'inner', on = 'EnrolNo', indicator= True)
merged
````

````python
merged = pd.merge(pers, marks, how = 'inner', on = 'EnrolNo',
                  suffixes= ('_pers', '_marks'),
                  indicator= True)
merged
````

````python
merged = pd.merge(pers, marks, how = 'inner', on = ['EnrolNo', 'Names'],
                  suffixes = ('_pers', '_marks'),
                  indicator= True)
merged
````

#### No matching columns

````python
pers = pd.DataFrame({'EnrolNo': ['A101', 'A102', 'A109', 'A107', 'A110', 'A112'],
                    'StudentNames': ['Aman', 'Shreay', 'Paul', 'Sam', 'John', 'Kia'],
                    'Interest' : ['Music', 'Arts', 'Music', 'Arts', 'Drama', 'Arts'],
                    'Age' : [12,13,12, 13,13, 12]})

marks = pd.DataFrame({'Id': ['A001', 'A102', 'A105', 'A109', 'A110'],
                      'Names' : ['Amy', 'Shreya', 'Mary', 'Paul', 'John'],
                     'English':[33, 44, 35, 37, 49],
                     'Math': [44,41,32, 30, 43] })
````

````python
merged = pd.merge(pers, marks, how = 'inner', left_on = 'EnrolNo', right_on = 'Id' )
merged
````

### <font size = 5 color = salmon ><b> more than 2 dataframes

````python
att = pd.DataFrame({'EnrolNo': ['A101', 'A109', 'A110'],
                   'att' : ['70%', '85%', '100%']})
````

````python
merged = pd.merge(pers, marks, how = 'outer', left_on = 'EnrolNo', right_on = 'Id' )
merged = merged[['EnrolNo', 'Names', 'Age', 'Interest', 'English', 'Math']]
merged
````

````python
merged = pd.merge(merged, att)
merged
````

#### pd.concat for merging

- axis = 0 ==> column wise operation : append (adding rows)   - we are adding values to each column
- axis = 1 ==> row wise operation    : merge (adding columns) - we are adding values to each row

**concat matches on index labels or column labels**

````python
merged = pd.concat([pers.set_index('EnrolNo'), marks.set_index('Id'), att.set_index('EnrolNo')],
         axis = 1, join= 'inner')
merged
````
